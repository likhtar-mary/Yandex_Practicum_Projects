# Описание проекта

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. 

**Задача**: создать модель для прогноза оттока клиентов и обучить ее на предоставленных данных.

**Описание услуг.**

Оператор предоставляет два основных типа услуг: 
1. Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: 
  - через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия»),
  - оптоволоконный кабель (Fiber optic).


Также доступны такие услуги:
1. Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
2. Выделенная линия технической поддержки (TechSupport);
3. Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
4. Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).


Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.

Во всех файлах столбец "customerID" содержит код клиента. Информация о договорах актуальна на 1 февраля 2020 года.

**План выполнения проекта:**
- Загрузка данных;
- Исследовательский анализ и предобработка данных;

- Выбор лучшей модели;
- Общий вывод и рекомендации заказчику.

Создаем функцию для построения графиков (гистограмма, диаграмма размаха) количественных непрерывных признаков:

```python
def viz_num_character(df, column, title, bins):
    sns.set(rc={'figure.figsize': (15, 7)})
    sns.set_palette('PRGn')
    fig, axes = plt.subplots(1, 2)
    sns.histplot(data=df, x=df[column], kde=True, bins=bins, ax=axes[0])
    axes[0].set(ylabel='Частота', xlabel=column)
    axes[0].set_title('Гистограмма распределения')
    median_value = df[column].median()
    mean_value = df[column].mean()
    axes[0].axvline(median_value, color='r', linestyle='--', label=f'Median: {median_value:.2f}')
    axes[0].axvline(mean_value, color='g', linestyle='-', label=f'Mean: {mean_value:.2f}')
    axes[0].legend()
    sns.boxplot(data=df[column], ax=axes[1])
    axes[1].set(ylabel=column)
    axes[1].set_title('Диаграмма размаха')
    plt.suptitle(title)
    plt.show()
```

Создаем функцию для построения графиков (столбчатая диаграмма) категориальных признаков и дискретных количественных:

### Добавление новых признаков

Согласно поставленной задачи, необходимо создать модель для прогнозирования оттока клиентов. Для этого необходимо выделить целевой признак "target": 0 - клиент продолжает пользоваться услугами, 1 - клиент ушел, расторгнул договор - задача классификации.

**Создаем целевой признак из признака "end_date"**, где:
- указанная дата расторжения договора означает, что клиент ушел - значение 1 в целевом признаке;
- значение "No" - клиент дальше пользуеся услугами - значение 0 в целевом признаке.

```python
# Функция для заполнения столбца с целевым признаком
def end_date (date):
    if date == 'No':
        return 0
    return 1
```

```python
# Применяем функцию и добавляем новый столбец в датафрейм
df_full['target'] = df_full['end_date'].apply(end_date)
```

## Подготовка данных

Делим данные на выборки: тренировочную и тестовую в соотношении 3:1. Используем стратификацию по целевому признаку (из-за дисбаланса классов):

```python
# Входные признаки
X = df_full.drop('target', axis=1)

# Целевой признак
y = df_full['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.25, 
    random_state=RANDOM_STATE, 
    stratify=y)

# Проверяем разделение на выборки:
print(f'''
Размерность X_train: {X_train.shape}
Размерность X_test: {X_test.shape}
Размерность y_train: {y_train.shape}
Размерность y_test: {y_test.shape}''')
```

Разделение прошло верно.

Создаем списки с названиями признаков:

```python
# Для кодирования 
cat_columns = ['type', 'paperless_billing', 'payment_method', 'senior_citizen', 'partner', 
               'dependents', 'internet_service', 'online_security', 'online_backup', 
               'device_protection', 'tech_support', 'streaming_tv', 'streaming_movies',
               'multiple_lines']

# Для масштабирования 
num_columns = ['total_charges', 'duration']
```

Признаки будут обработаны таким образом:
1. для логистической регрессии:
  - OneHotEncoder для кодирования категориальных признаков;
  - MinMaxScaler для масштабирования количественных признаков;
2. для дерева решений и градиентного бустинга:
  - OrdinalEncoder для кодирования категориальных признаков;
  - MinMaxScaler для масштабирования количественных признаков.
  
Собираем трансформеры:

```python
# Для модели логистической регрессии
preprocessor_lr = ColumnTransformer(
    [
        ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False), cat_columns),
        ('num', MinMaxScaler(), num_columns)
    ], remainder='passthrough'
)

preprocessor_lr
```

```python
# Для моделей дерева решений и градиентного бустинга
preprocessor_tr = ColumnTransformer(
    [
        ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), cat_columns),
        ('num', MinMaxScaler(), num_columns)
    ], remainder='passthrough'
)

preprocessor_tr
```

Шаги подготовки данных и инициализации модели будут на следующем шаге "Обучение моделей".

**Вывод:**

**На этапе подготовки данных было выполнено:**

1. **данные разделены на выборки**: тренировочную и тестовую в соотношении 3:1. Применена стратификация по целевому признаку (из-за дисбаланса классов);


2. **подготовка данных**:
  - созданы трансформеры, которые включают в себя шаги кодирования категориальных признаков с помощью OneHotEncoder (логистическая регрессия) и OrdinalEncoder (дерево решений и градиентный бустинг), масштабирования (стандартизации количественных признаков).
  - итоговый пайплайн будет включать в себя шаги подготовки данных и инициализацию модели на следующем шаге "Обучение моделей".

## Обучение моделей машинного обучения

### Модель LogisticRegression

Cоздаем итоговый пайплайн pipe_final, включающий в себя шаги подготовки данных и инициализацию модели логистической регрессии, находим лучшие гиперпараметры модели с помощью автоматизированого поиска GridSearchCV, в качестве метрики используем ROC-AUC:

```python
# Итоговый пайплайн
pipe_final= Pipeline([
    ('preprocessor', preprocessor_lr),
    ('lr', LogisticRegression(random_state=RANDOM_STATE))
])

# Словарь для модели LogisticRegression()
param_grid_lr = {'lr__solver': ('liblinear', 'saga','newton-cg', 'lbfgs'),
                 'lr__C': (.1, 1, 5, 10, 15),
                 'lr__max_iter': ([500]),
                 'lr__class_weight': ([None, 'balanced'])}

# Запускаем поиск гиперпараметров
grid_search_lr = GridSearchCV(
    pipe_final,
    param_grid_lr,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)
grid_search_lr.fit(X_train, y_train)
```

```python
print('Лучшая модель и её параметры LogisticRegression:', grid_search_lr.best_params_)
print('Лучшая метрика ROC-AUC при кросс-валидации:', grid_search_lr.best_score_.round(3))
```

Сохраняем лучшую обученную модель логистической регрессии:

```python
best_lr = grid_search_lr.best_estimator_
```

**Вывод**: обучена модель логистической регрессии и найдены лучшие гиперпараметры с помощью автоматизированого поиска GridSearchCV, в качестве метрики выбрана ROC-AUC.

**Выбрана лучшая модель с гиперпараметрами: LogisticRegression (C=15, max_iter=500, random_state=280725, solver='newton-cg', class_weight='balanced')**.
